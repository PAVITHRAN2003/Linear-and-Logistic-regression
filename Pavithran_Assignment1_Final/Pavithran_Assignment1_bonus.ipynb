```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv("Crime_Incidents_20241010.csv")
```

```python


# Identify the index of the 'Longitude' column
longitude_index = df.columns.get_loc('Longitude')

# Keep only the columns up to and including 'Longitude'
df = df.iloc[:, :longitude_index + 1]

df.drop(columns=['Incident ID'], inplace=True)
print(df.info())
print(df.isnull().sum())
```

```python
df= df.dropna()
```

```python
print(df.head(15))
print(df.shape)
```

```python
df.drop(columns=['Longitude','Incident Datetime', 'Latitude', 'Case Number', 'Location'], inplace=True)
```

```python
df.head()
```

```python
df['Address'] = df['Address'].str.extract(r'([A-Z]+\s[A-Z]+)')[0]  # Adjust regex based on your address format

categorical_columns = ['Incident Type Primary','Incident Description', 'Parent Incident Type', 'Day of Week', 'City', 'State', 'Address']
for column in categorical_columns:
    df[column] = df[column].astype('category').cat.codes
df['Hour of Day'] = df['Hour of Day'].apply(lambda x: 1 if x < 8 else 0)





```

```python

Y =df['Hour of Day'].values
X = df.drop(columns=['Hour of Day'])

# X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
import random
N = X.shape[0]
indices = list(range(N))
random.seed(123)
random.shuffle(indices)

train_size = int(0.8 * N)
train_indices = indices[:train_size]
test_indices = indices[train_size:]

X_train = X.iloc[train_indices]
Y_train = Y[train_indices]
X_test = X.iloc[test_indices]
Y_test = Y[test_indices]

print("Shape of X_train: ",X_train.shape,"Shape of X_test: ",X_test.shape)
print("Shape of Y_train: ",Y_train.shape,"Shape of Y_test: ",Y_test.shape)
```

```python

class LogitRegression():
    def __init__(self, learning_rate=0.005, iterations=5000):
        self.learning_rate = learning_rate
        self.iterations = iterations

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def cost(self, h, y):
        m = len(y)
        return (-1 / m) * np.sum(y * np.log(h + 1e-10) + (1 - y) * np.log(1 - h + 1e-10))

    def gradient_descent(self, X, y, h):
        m = len(y)
        delta = h - y.flatten()
        dW = (1 / m) * np.dot(X.T, delta)
        db = (1 / m) * np.sum(delta)
        self.w -= self.learning_rate * dW
        self.b -= self.learning_rate * db

    def fit(self, X, y):
        X = np.array(X, dtype=float)
        y = np.array(y, dtype=float).flatten()

        assert X.ndim == 2
        assert y.ndim == 1
        assert X.shape[0] == y.shape[0]

        self.w = np.random.uniform(0, 1, X.shape[1])
        self.b = 0
        self.losses = []

        for i in range(self.iterations):
            z = np.dot(X, self.w) + self.b
            h = self.sigmoid(z)

            loss = self.cost(h, y)
            self.losses.append(loss)

            self.gradient_descent(X, y, h)

            if i % 100 == 0:
                print(f'Iteration {i}, Loss: {loss}')

    def predict(self, X):
        z = np.dot(X, self.w) + self.b
        h = self.sigmoid(z)
        return np.where(h >= 0.5, 1, 0)

    def save_weights(self, file_name="model_weights.pkl"):
        with open(file_name, 'wb') as file:
            pickle.dump({'weights': self.w, 'bias': self.b}, file)
        print(f"Weights and bias saved to {file_name}")
```

```python
model = LogitRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
```

```python
# print(y_pred[:10])
# print(Y_test[:10])
correct_predictions = np.sum(y_pred == Y_test)

accuracy = correct_predictions / len(Y_test) * 100
print(accuracy)
model.save_weights('Pavithran_Assignment1_bonus.pkl')
```
